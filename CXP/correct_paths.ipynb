```python
import pandas as pd
```

```python
test_df = pd.read_csv('Dataframes/preprocessed_test_CXP.csv')
test_df.head()
```

```python
train_df = pd.read_csv('Dataframes/preprocessed_train_CXP.csv')
val_df = pd.read_csv('Dataframes/preprocessed_val_CXP.csv')
```

```python
img_path_CXP = "/bigdata/andrei_thesis/CXP_data/images/"
```

```python
def add_image_path_CXP(path):
    if not isinstance(path, str):
        return ""
    return img_path_CXP+"/".join(path.split('/')[2:])
```

```python
new_test_paths = test_df['path'].apply(add_image_path_CXP)
test_df['path'] = new_test_paths
test_df.rename(columns={'path':'Jointpath'}, inplace=True)
test_df.head()
```

```python
print(test_df['Jointpath'][0])
```

```python
new_train_paths = train_df['path'].apply(add_image_path_CXP)
new_val_paths = val_df['path'].apply(add_image_path_CXP)
train_df['path'] = new_train_paths
train_df.rename(columns={'path':'Jointpath'}, inplace=True)
val_df['path'] = new_val_paths
val_df.rename(columns={'path':'Jointpath'}, inplace=True)
```

```python
display(train_df.head())
display(val_df.head())
```

```python
test_df.to_csv('preprocessed_test_CXP_mevis.csv', index=False)
train_df.to_csv('preprocessed_train_CXP_mevis.csv', index=False)
val_df.to_csv('preprocessed_val_CXP_mevis.csv', index=False)
```
