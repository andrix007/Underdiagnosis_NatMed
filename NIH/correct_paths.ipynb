```python
import pandas as pd
```

```python
test_df = pd.read_csv('preprocessed_test_df.csv')
test_df.head()
```

```python
train_df = pd.read_csv('preprocessed_train_df.csv')
val_df = pd.read_csv('preprocessed_val_df.csv')
```

```python
img_path_NIH = "/bigdata/andrei_thesis/NIH_data/images/images/"
```

```python
def add_image_path_NIH(path):
    return img_path_NIH+path
```

```python
new_test_paths = test_df['path'].apply(add_image_path_NIH)
test_df['path'] = new_test_paths
test_df.rename(columns={'path':'Jointpath'}, inplace=True)
test_df.head()
```

```python
new_train_paths = train_df['path'].apply(add_image_path_NIH)
new_val_paths = val_df['path'].apply(add_image_path_NIH)
train_df['path'] = new_train_paths
train_df.rename(columns={'path':'Jointpath'}, inplace=True)
val_df['path'] = new_val_paths
val_df.rename(columns={'path':'Jointpath'}, inplace=True)
```

```python
display(train_df.head())
display(val_df.head())
```

```python
test_df.to_csv('preprocessed_test_df_mevis.csv', index=False)
train_df.to_csv('preprocessed_train_df_mevis.csv', index=False)
val_df.to_csv('preprocessed_val_df_mevis.csv', index=False)
```
