```python
import pandas as pd
```

```python
test_df = pd.read_csv('data/preprocessed_mimic_test.csv')
test_df.head()
```

```python
train_df = pd.read_csv('data/preprocessed_mimic_train.csv')
val_df = pd.read_csv('data/preprocessed_mimic_valid.csv')
```

```python
img_path_MIMIC = "/bigdata/andrei_thesis/MIMIC/mimic-cxr-2.0.0.physionet.org/"
```

```python
def add_image_path_MIMIC(path):
    if not isinstance(path, str):
        return ""
    return img_path_MIMIC+path
```

```python
new_test_paths = test_df['path'].apply(add_image_path_MIMIC)
test_df['path'] = new_test_paths
test_df.rename(columns={'path':'Jointpath'}, inplace=True)
test_df.head()
```

```python
new_train_paths = train_df['path'].apply(add_image_path_MIMIC)
new_val_paths = val_df['path'].apply(add_image_path_MIMIC)
train_df['path'] = new_train_paths
train_df.rename(columns={'path':'Jointpath'}, inplace=True)
val_df['path'] = new_val_paths
val_df.rename(columns={'path':'Jointpath'}, inplace=True)
```

```python
display(train_df.head())
display(val_df.head())
```

```python
test_df.to_csv('preprocessed_test_MIMIC_mevis.csv', index=False)
train_df.to_csv('preprocessed_train_MIMIC_mevis.csv', index=False)
val_df.to_csv('preprocessed_val_MIMIC_mevis.csv', index=False)
```
